````python
# Importando bibliotecas
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import pyplot
import plotly.express as px
import seaborn as sns
np.random.seed(2360873)
N = 100
````

````python
# Carregando base de dados
df = pd.read_csv('previsao_de_renda.csv')

df1 = df.drop(df[['Unnamed: 0','data_ref', 'index']], axis=1)
df1.head()
````

# **Pessoas com imóveis**
---

````python
sns.countplot(x='posse_de_imovel', data=df1)
````

# **Pessoas com veículos**

````python
sns.countplot(x='posse_de_veiculo', data=df1)
````

# **Histrograma Geral**

````python
sns.displot(data = df1
,x = 'renda'
,bins = 30            
,element = 'step'
)
````

# **Histograma - Posse de imóvel**

````python
sns.displot(data = df1
,x = 'renda'
,hue = 'posse_de_imovel'
,bins = 30            
,element = 'step'
,height = 4.5
            
)
````

Pelos histogramas a distribuição da renda das pessoas com imóvel e sem imóvel são bem semelhantes, levando-se em conta que há mais pessoas sem imóveis nos dados. Não sendo possível notar a influência deste atributo na renda.

# **Histograma - Posse de veículo**

````python
sns.displot(data = df1
,x = 'renda'
,hue = 'posse_de_veiculo'
,bins = 30
,element = 'step'
)
````

Pelos histogramas a distribuição da renda das pessoas com veículo e sem veículo são bem semelhantes, levando-se em conta que há mais pessoas sem veículos nos dados. Não sendo possível notar a influência deste atributo na renda.

# **Perfis de médias**

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=df1, dodge=True, ci = 90)
ax.set_title('Perfis Media Renda - Posse de Imóvel')
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=df1, dodge=True, ci = 90)
ax.set_title('Perfis Media Renda - Posse de Veículo')
````

Pelos perfis de média podemos ver que as faixas de médias da renda das amostras de quem tem imóveis e quem tem veículo são bastante superiores das que não tem. Isso pode indicar que estes atributos são bons preditores da renda das pessoas.
